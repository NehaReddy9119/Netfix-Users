<h2> Importing required libraries </h2>
Here , I am importing libraries required i.e Pandas,Matplotlib and seaborn

```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

<h2>Reading our Dataset</h2>
The Dataset being used here is Netflix User Database . It contains the user information having columns named Subscription_type , Monthly Revenue , Join Date,Last Payment Date,Country,Age,Gender,Device,Plan Duration.
We do various data cleaning and data analytic operations on our above mentioned dataset.

```python
df = pd.read_csv(r"C:\Users\dell\Downloads\Netflix Userbase.csv")
df
```

```python
df.shape
```

```python
df.info()
```

Setting the index in the dataset to User ID rather than having an extra index column in the dataset

```python
df.set_index('User ID' ,inplace = True)
df
```

Renaming our index column name to User_id

```python
df = df.rename_axis('User_id')
df
```

Renaming all the column names

```python
df.columns = [x.replace(" ","_") for x in df.columns]
df
```

Checking if there are any null values present in our database

```python
df.isna().sum()
```

Changing the datatype of the date columns present from string to dateformat

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'],format = '%d-%m-%y')
df['Last_Payment_Date'] = pd.to_datetime(df['Last_Payment_Date'],format = '%d-%m-%y')
```

<h3>Finding how many months it has been since the last payment of users</h3>

```python
current_date = datetime.now()
df['Months_since_LastPayment'] = (current_date - df['Last_Payment_Date']).dt.days // 30
df
```

```python
df['Months_since_LastPayment'].value_counts()
```

```python
date_difference = current_date - df['Last_Payment_Date']
months = date_difference.dt.days // 30
days = date_difference.dt.days % 30
df['Days_since_LastPayment'] = days
df
```
